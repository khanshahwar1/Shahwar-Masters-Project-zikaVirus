```python
import pandas as pd
zika_data = pd.read_csv("zika-data.csv")
zika_data
zika_data["human.pid"] = zika_data["human.pid"].map(lambda x: x[10:])
zika_data["zika.pid"] = zika_data["zika.pid"].map(lambda x: x[10:])
zika_data
zika_data=zika_data.rename(columns={'zika_sequence':'pathogen_sequence'})
```

```python
zika_data
```

```python
zika_data[zika_data.duplicated()]
zika_data = zika_data.drop_duplicates(keep=False)
zika_data.drop(zika_data.tail(1).index,inplace=True) 
zika_data
```

```python
zika_data['length']= zika_data['human_sequence'].apply(len)
zika_data.drop(zika_data[zika_data.length < 35].index, inplace=True)
zika_data
```

```python
joinData = zika_data["human_sequence"]+zika_data["pathogen_sequence"]
humanData = zika_data["human_sequence"]
zikaD = zika_data["pathogen_sequence"]
print(joinData)
print("\n")
print("\n")
print(humanData)
print("\n")
print("\n")
print(zikaD)
```

```python
arrayList_A = []
for i,j in joinData.items():
    Result = []
    Result = [(j[i:i+1]) for i in range(0,len(j),1)]
    arrayList_A.append(Result)
```

```python
arrayList_B = []
for i,j in humanData.items():
    Result = []
    Result = [(j[i:i+1]) for i in range(0,len(j),1)]
    arrayList_B.append(Result)
```

```python
arrayList_C = []
for i,j in zikaD.items():
    Result = []
    Result = [(j[i:i+1]) for i in range(0,len(j),1)]
    arrayList_C.append(Result)
```

```python
columnA = pd.Series(arrayList_A)
columnB = pd.Series(arrayList_B)
columnC = pd.Series(arrayList_C)
dataA = pd.DataFrame({'Human':columnB,'Pathogen':columnC,'Joined':columnA})
dataA = dataA.sample(frac=1)
dataA = dataA.reset_index()
dataA = dataA.drop('index', 1)
dataA
```

```python
count1 = dataA.shape[0]
if count1 % 2 == 0:
    one=zero=int(count1/2)
else:
    one = int(count1/2)+1
    zero= one-1
labelList = [1]*one+[0]*zero
```

```python
import random
random.shuffle(labelList)
dataA['label'] = labelList
dataA
```

```python
import pickle
dataA.to_pickle("formatted.pkl")
dataFinal = pd.read_pickle("formatted.pkl")
print(dataFinal.head(15))
dataset_train = dataFinal[0:3000]
dataset_test = dataFinal[3001:]
dataset_test = dataset_test.reset_index()
dataset_test = dataset_test.drop('index', 1)
```
